```python
import pandas as pd

df = pd.read_csv("optimizer_results.csv")

df.head()
```

```python
import numpy as np

np.random.seed(42)

df["Latitude"] = np.random.uniform(
    18.40, 18.80, len(df)
)

df["Longitude"] = np.random.uniform(
    73.70, 74.10, len(df)
)
```

```python
top_atms = (
    df.sort_values("Priority_Rank")
      .drop_duplicates(subset="ATM_ID")
      .head(10)
      .reset_index(drop=True)
)

top_atms[
    [
        "ATM_ID",
        "Recommended_Refill",
        "Risk_Level",
        "Priority_Rank"
    ]
]
```

```python
import numpy as np

np.random.seed(42)

top_atms["Latitude"] = np.random.uniform(
    18.40, 18.80, len(top_atms)
)

top_atms["Longitude"] = np.random.uniform(
    73.70, 74.10, len(top_atms)
)

top_atms[
    [
        "ATM_ID",
        "Latitude",
        "Longitude"
    ]
]
```

```python
from scipy.spatial.distance import cdist
import pandas as pd

# Create coordinate array
coordinates = top_atms[["Latitude", "Longitude"]].values

# Compute distance matrix
distance_matrix = cdist(
    coordinates,
    coordinates,
    metric="euclidean"
)

# Convert to DataFrame for readability
distance_df = pd.DataFrame(
    distance_matrix,
    index=top_atms["ATM_ID"],
    columns=top_atms["ATM_ID"]
)

distance_df.round(2)
```
